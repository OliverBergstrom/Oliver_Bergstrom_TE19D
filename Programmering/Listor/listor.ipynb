# Listor

```python
# indexering

frukter = ["äpple", "banan", "päron", "jordgubbe", "hallon", "kiwi"]
print(f"frukter[1]: {frukter[1]}")
print(f"frukter[-1]: {frukter[-1]}")
# skivoperator
print(f"frukter[1:3]: {frukter[1:3]}")

print("\n")

# loopa igenom listan:

for frukt in frukter:
    print(f"min fruktkorg innehåller: {frukt}")
```

```python
frukter = ["äpple", "banan", "päron", "jordgubbe", "hallon", "kiwi"]
grönsaker = ["sallad", "gurka", "tomat"]

fruktsallad = frukter

for grönsak in grönsaker:
    fruktsallad.append(grönsak)

print(fruktsallad)
```

```python
import matplotlib.pyplot as plt
# list comprehension
y = [2*x-2 for x in range(10)]
x = [x for x in range(10)]
print(f"y: {y}")
print(f"x: {x}")

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graf till y=2x-2")
plt.show()
```
