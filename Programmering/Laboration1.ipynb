```python
# e), f), g), h)

import random # importerar kod som tillåter mig att göra slumpmässiga kommandon
import math # importerar matterelaterade kommandon
import numpy as np # importerar kommandon som tillåter mig att göra listor
import matplotlib.pyplot as plt # importerar komandon som tillåter mig att göra en graf

i = 0 # startvärde för hur många punkter som finns

for j in range(5000): # hur många punkter som ska simuleras
    x = random.uniform(-1,1) # simulerar x punkter
    y = random.uniform(-1,1) # simulerar y punkter

    avstånd = np.sqrt(x**2+y**2) # max avståndet en punkt innanför cirkeln kan vara

    if avstånd < 1: # om avståndet är mindre än ett så är punkten innanför cirkeln
        i += 1 # lägger till antalet punkter innanför cirkeln
        plt.plot(x,y,'bo')
    else:
        plt.plot(x,y,'ro')

plt.show()

k = 4*i/5000 # antal punkter innanför gånger fyra sen delat på totala antalet punkter simulerade

print(k) # desto mer punkter man simulerar, ju närmare kommer man pi
```

```python
# a)
for i in range(1,101): # start och slutvärdet i räknaren
    print(i, end=" ") # skriver ut räknaren och ett mellanslag efter
```

```python
# b)
for i in range(1,101):
    if i%5 == 0: # kollar om talet som räknas ut går att dela på 5, om den gör det så blir i%5 = 0. Om det blir 0 så printar den burr
        print("burr")   
    else:   # Om inte så fortsätter den printa ut nummrena
        print(i, end=" ")   
```

```python
# c)
burr = int(input("Ange din multipel")) # gör så att användaren kan mata in ett tal som blir multipeln

for i in range(1,101):
    if i%burr: # kollar om talet går att dela med multipeln som användaren matade in
        print(i, end=" ")
    else: 
        print("burr")
```

```python
# d)
burr = int(input("Ange din multipel")) # användaren matar in sitt valda burr
birr = int(input("Ange din multipel")) # användaren matar in sitt valda birr

for i in range(1,101):
    if i%burr==0: # kollar så att den kan delas på det valda burr
        print("burr", end=" ")
    
    elif i%birr==0: # kollar så att den kan delas på det valda birr
        print("birr", end=" ")
    
    else:
        print(i, end=" ")


    
```

```python
# e)
start = int(input("Ange startvärde")) # Låter användaren mata in räknarens startvärde
slut = int(input("Ange slutvärde")) + 1 # Låter användaren mata in räknarens slutvärde

burr = int(input("Ange din multipel")) # Låter användaren mata in sitt burr
birr = int(input("Ange din multipel")) # Låter användaren mata in sitt birr

aburr = 0 # Startvärdet för antalet burr
abirr = 0 # Startvärdet för antalet birr

for i in range(start,slut):
    if i%burr==0:   # Ifall nummret är delbart med burr så blir det +1 i burr räknaren
        print("burr", end=" ") 
        aburr += 1
    
    elif i%birr==0: # Ifall nummret är delbart med birr så blir det +1 i birr räknaren
        print("birr", end=" ") 
        abirr += 1
    
    else:
        print(i, end=" ")


print(f"antalet birr:{abirr}")  # Printar antalet birr
print(f"antalet burr:{aburr}")  # Printar antalet burr
        
```
